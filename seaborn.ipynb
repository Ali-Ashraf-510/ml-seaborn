```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df_tips  = sns.load_dataset('tips')
# داتا سيت جاهزة جوا المكتبة للتعلم
```

```python
df_tips.head()
```

```python
# بنحدد حجم الرسم البياني عشان يبقى 10x6
plt.figure(figsize=(10,6))

# بنرسم scatter plot باستخدام مكتبة seaborn عشان نشوف العلاقة بين 'total_bill' و 'tip'
# وبنستخدم 'hue' عشان نميز بين الجنسين (sex) بألوان مختلفة
sns.scatterplot(x='total_bill', y='tip', data=df_tips, hue='sex' , s=100)

plt.xlabel('Total Bill (in dollars)', fontsize=12)

# نكتب عنوان أوضح للمحور الرأسي
plt.ylabel('Tip (in dollars)', fontsize=12)

# بنغير عنوان الرسم البياني
plt.title('Relationship Between Total Bill and Tip by Gender', fontsize=16)

plt.grid(ls = '--' )
# بنعرض الرسم البياني
plt.show()
```

```python
# بنرسم علاقة بين 'total_bill' و 'tip' باستخدام relplot
# وبنستخدم نوع الرسم "scatter" عشان يبقى نقاط
# col='sex' بيوزع الرسم في أعمدة على حسب الجنس (ذكر أو أنثى)
# row='smoker' بيوزع الرسم في صفوف على حسب حالة التدخين (مدخن أو غير مدخن)
sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', col='sex', row='smoker')

# بنعرض الرسم البياني
plt.show()
```
